```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
ip = iris.data[:,2:]
op = iris.target
```

```python
plt.scatter(ip[:,0],ip[:,1],c=op)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(ip,op)
```

```python
x_max,x_min = ip[:,0].max()+1, ip[:,0].min()-1
y_max,y_min = ip[:,1].max()+1,ip[:,1].min()-1

xx,yy = np.meshgrid(np.linspace(x_min,x_max),
                   np.linspace(y_min,y_max))
grid = np.c_[xx.ravel(),yy.ravel()]
pred = model.predict(grid).reshape(xx.shape)
plt.contourf(xx,yy,pred)
plt.scatter(ip[:,0],ip[:,1],c=op)
plt.title('logistic')
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB

modelNB = GaussianNB()
modelNB.fit(ip,op)
```

```python
predNB = modelNB.predict(grid).reshape(xx.shape)

plt.contourf(xx,yy,predNB)
plt.scatter(ip[:,0],ip[:,1],c=op)
plt.title('bayes')
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
modelSV = SVC(kernel ='poly',C=100,gamma=0.1)
```

```python
modelSV.fit(ip,op)
```

```python
predSV = modelSV.predict(grid).reshape(xx.shape)

plt.contourf(xx,yy,predSV)
plt.scatter(ip[:,0],ip[:,1],c=op)
plt.title('SVC')
plt.show()
```
